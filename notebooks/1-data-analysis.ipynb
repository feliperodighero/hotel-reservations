### ðŸ‡ºðŸ‡¸ Correlation Matrix - ðŸ‡§ðŸ‡· Matrix de CorrelaÃ§Ã£o

ðŸ‡ºðŸ‡¸ We can see that there isn't a strong correlation between the variables. The strongest correlation is between repeated_guest and no_of_previous_bookings_not_canceled.

ðŸ‡§ðŸ‡· A gente pode perceber que nÃ£o existe uma correlaÃ§Ã£o forte entre as variÃ¡veis. A maior correlaÃ§Ã£o Ã© entre as variÃ¡veis repeated_guest e no_of_previous_bookings_not_canceled.

```python
correlation_matrix = df.corr(numeric_only=True)

plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})
plt.title('Matrix de correlaÃ§Ã£o entre variÃ¡veis numÃ©ricas', fontsize=16)
plt.tight_layout()
# plt.savefig('../reports/figures/correlation_matrix.png', dpi=300)
plt.show()
```

```python
columns = [
    'no_of_adults', 'no_of_children', 'no_of_weekend_nights',
    'no_of_week_nights', 'required_car_parking_space', 'room_type_reserved',
    'arrival_year', 'arrival_month', 'repeated_guest',
    'no_of_previous_cancellations', 'no_of_special_requests'
]

plt.figure(figsize=(20, 25))

for i, col in enumerate(columns, 1):
    plt.subplot(6, 2, i)
    plt.gca().set_title(f'Variable {col}')
    sns.countplot(x=col, palette='Set2', data=df)

plt.title('DistribuiÃ§Ã£o das variÃ¡veis categÃ³ricas', fontsize=16)
plt.tight_layout()
# plt.savefig('../reports/figures/plot-variables-1.png', dpi=300)
plt.show()
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette("husl")

fig = plt.figure(figsize=(24, 28))
fig.patch.set_facecolor('#f8f9fa')

fig.suptitle('AnÃ¡lise Detalhada do Status de Reservas\nDistribuiÃ§Ã£o por Diferentes VariÃ¡veis',
             fontsize=28, fontweight='bold', y=0.95, color='#2c3e50')

custom_colors = ['#3498db', '#e74c3c', '#2ecc71', '#f39c12', '#9b59b6', '#1abc9c', '#34495e', '#e67e22']

variables = [
    ('no_of_adults', 'NÃºmero de Adultos'),
    ('no_of_children', 'NÃºmero de CrianÃ§as'),
    ('no_of_weekend_nights', 'Noites de Fim de Semana'),
    ('market_segment_type', 'Segmento de Mercado'),
    ('type_of_meal_plan', 'Tipo de Plano de RefeiÃ§Ã£o'),
    ('required_car_parking_space', 'Necessidade de Estacionamento'),
    ('room_type_reserved', 'Tipo de Quarto Reservado'),
    ('arrival_year', 'Ano de Chegada')
]

for i, (var, title) in enumerate(variables, 1):
    ax = plt.subplot(4, 2, i)

    if var in df.columns:
        cross_tab = pd.crosstab(df['booking_status'], df[var], normalize='index') * 100

        sns.countplot(data=df, x='booking_status', hue=var,
                     palette=custom_colors[:len(df[var].unique())],
                     ax=ax, alpha=0.8)

        for container in ax.containers:
            ax.bar_label(container, fmt='%d', fontsize=9, fontweight='bold')

    ax.set_title(title, fontsize=16, fontweight='bold', pad=20, color='#2c3e50')
    ax.set_xlabel('Status da Reserva', fontsize=12, fontweight='bold', color='#34495e')
    ax.set_ylabel('Quantidade', fontsize=12, fontweight='bold', color='#34495e')

    ax.tick_params(axis='x', labelsize=10, rotation=45)
    ax.tick_params(axis='y', labelsize=10)
    ax.grid(True, alpha=0.3, linestyle='--')

    if ax.get_legend():
        ax.legend(title=title.split(' ', 1)[1], title_fontsize=10,
                 fontsize=9, loc='upper right', framealpha=0.9,
                 fancybox=True, shadow=True)

    for spine in ax.spines.values():
        spine.set_edgecolor('#bdc3c7')
        spine.set_linewidth(1.5)

plt.tight_layout(rect=[0, 0.03, 1, 0.92])

fig.text(0.5, 0.02,
         f'Total de registros analisados: {len(df):,} | '
         f'VariÃ¡veis analisadas: {len(variables)} | '
         f'Gerado automaticamente',
         ha='center', fontsize=12, style='italic', color='#7f8c8d')

# plt.savefig('../reports/figures/Analysis of variable bookin_status 1.png', dpi=300)
plt.show()

fig2, axes = plt.subplots(4, 2, figsize=(20, 24))
fig2.patch.set_facecolor('#f8f9fa')
fig2.suptitle('AnÃ¡lise Proporcional do Status de Reservas\nDistribuiÃ§Ã£o Percentual por VariÃ¡vel',
              fontsize=24, fontweight='bold', y=0.95, color='#2c3e50')

for i, (var, title) in enumerate(variables):
    row, col = i // 2, i % 2
    ax = axes[row, col]

    if var in df.columns:
        cross_tab = pd.crosstab(df[var], df['booking_status'], normalize='index') * 100

        cross_tab.plot(kind='bar', ax=ax, stacked=True,
                      color=['#e74c3c', '#2ecc71'], alpha=0.8)

        for container in ax.containers:
            ax.bar_label(container, fmt='%.1f%%', fontsize=8,
                        label_type='center', fontweight='bold', color='white')

    ax.set_title(title, fontsize=14, fontweight='bold', pad=15, color='#2c3e50')
    ax.set_xlabel(title.split(' ', 1)[1], fontsize=11, fontweight='bold')
    ax.set_ylabel('Percentual (%)', fontsize=11, fontweight='bold')
    ax.tick_params(axis='x', rotation=45, labelsize=9)
    ax.tick_params(axis='y', labelsize=9)
    ax.grid(True, alpha=0.3, axis='y')
    ax.legend(title='Status', loc='upper right', fontsize=9)

plt.tight_layout(rect=[0, 0.03, 1, 0.92])
# plt.savefig('../reports/figures/Analysis of variable bookin_status 2.png', dpi=300)
plt.show()
```
